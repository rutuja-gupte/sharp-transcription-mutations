```python
from pypdf import PdfReader

reader = PdfReader("muller_supp.pdf")
text = ""
pages = reader.pages[42:46]
```

```python
page42 = reader.pages[42].extract_text()
page43 = reader.pages[43].extract_text()
page44 = reader.pages[44].extract_text()
page45 = reader.pages[45].extract_text()[:442]
```

```python
data = []
data.extend(re.findall(r"(ARS\d+?\.?\d?)\t\r \xa0(.*?)\t\r \xa0", page42))
data.extend(re.findall(r"(ARS\d+?\.?\d?)\t\r \xa0(.*?)\t\r \xa0", page43))
data.extend(re.findall(r"(ARS\d+?\.?\d?)\t\r \xa0(.*?)\t\r \xa0", page44))
data.extend(re.findall(r"(ARS\d+?\.?\d?)\t\r \xa0(.*?)\t\r \xa0", page45))
```

```python
ori = pd.read_csv("yeastmine_results.csv")
ori = ori.set_axis(['alias', 'loc', 'chr', 'name', 'name2'], axis=1)
ori.head()
```

```python
yays = []
nahs = []
```

```python
for d in data:
    if d[0] in list(ori["name"]) or d[0] in (ori["name2"]):
        yays.append(d)
    else:
        nahs.append(d)
```

```python
alias = ori["alias"]
aliases= []

for i in alias.str.split("|"):
    try:
        aliases.extend(i)
    except TypeError:
        aliases.append(i)
```

```python
new_nahs = []
for nah in nahs:
    if nah[0] in aliases:
        print(nah)
        yays.append(nah)
        new_nahs.append(nah)
```

```python
for nah in new_nahs:
    nahs.pop(nahs.index(nah))
```

```python
nahs
```

```python
temp_df = alias.str.split("|", expand=True)
df = pd.concat([ori, temp_df], axis=1)
df
```

```python
df.columns
```

```python
l = []
for yay in yays:
    d = {}
    small_df = df[(df["name"] == yay[0]) | (df["name2"] == yay[0]) | \
       (df[0] == yay[0]) | (df[1] == yay[0]) | \
       (df[2] == yay[0])].iloc[0]
    d["name"] = yay[0]
    d["seq"] = yay[1]
    d["chr"] = small_df["chr"]
    d["loc"] = small_df["loc"]
    l.append(d)
```

```python
final_df = pd.DataFrame(l)
final_df
```

```python
final_df.to_csv("muller_rep.csv")
```

```python
table = soup.find("table")
rows = table.find_all("tr")
t = []
for row in rows:
    r = [item.text for item in row.find_all("td")]
    t.append(r)
```
