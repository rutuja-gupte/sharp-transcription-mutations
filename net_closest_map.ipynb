```python
feat["reads"] = 0
net["type"] = None
feat = feat.rename(columns = {"class":"type"})
```

```python
# feat_test = feat.sort_values(by = ["strand", "chr_num", "left"]).iloc[15:20]
feat_test = feat.sort_values(by = ["strand", "chr_num", "left"])
# feat_test = feat_test.set_index(["strand", "chr_num", "left", "right"])
feat_test
```

```python
# d_test = net.iloc[800:805]
d_test = net.iloc[:10000]
d_test
```

```python
t0 = time.time()
i = 0
for d in d_test.itertuples():
#     l_tuples = []
#     for t in feat_test.itertuples():
#         if t.strand == d.strand and t.chr_num == d.chr_num:
#             if t.left < d.pos and t.right > d.pos:
#                 l_tuples.append(t)
#     f = sorted(l_tuples, key=lambda item: d.pos-item.left)[0]

    if i%1000 == 0:
        print(time.time() - t0)
        
    condition1 = (feat_test["strand"] == d.strand)
    condition2 = (feat_test["chr_num"] == d.chr_num) 
    condition3 = (feat_test["left"] < d.pos)
    condition4 = (feat_test["right"] > d.pos)
    df = feat_test[condition1 & condition2 & condition3 & condition4]


#     df = feat_test[(feat_test["strand"] == d.strand) & \
#              (feat_test["chr_num"] == d.chr_num) & \
#              (feat_test["left"] < d.pos) & \
#              (feat_test["right"] > d.pos)]

    if len(df) != 0:
        dists = (d.pos - df["left"])
        location = dists.argmin()
        feat_test.at[dists.index[location], "reads"] += d.reads
    
    i += 1
```
